# 1. Background - Ensemble Algorithmen

<u>**Ensemble Algorithmen:**</u>

•	Das Ziel von Ensemble-Methoden ist es, die Vorhersagen mehrerer Base-Estimator, die mit einem bestimmten Lernalgorithmus erstellt wurden, zu kombinieren, um die Generalisierbarkeit/Robustheit gegenüber einem einzelnen Estimator zu verbessern.

Arten von Ensemble Algorithmen **Bagging**, **Boosting** und Stacking

Wir können zwischen parallelen und sequentiellen Ensemble-Methoden sowie heterogenen und homogenen Ensemble-Methoden entscheiden.

<u>**Bagging:**</u>

Bagging (Bootstrap Aggregation) ist eine einfache und sehr leistungsfähige Ensemble-Methode für Klassifikations- und Regressionsaufgaben. Bagging ist ein allgemeines Verfahren, das zur Verringerung der Varianz bei Algorithmen mit hoher Varianz verwendet werden kann. Ein Algorithmus, der eine hohe Varianz aufweist, ist der Entscheidungsbaum, wie Klassifizierungs und Regressionsbäume (CART). Ein weiteres Beispiel für einen Bagging-Algorithmus ist der Random Forest Algorithmus.

Bootstraping

Im Wesentlichen handelt es sich beim Bootstrapping um eine statistische Methode, bei der Zufallsstichproben (Subsamples) mit Zurücklegen (d.h. wir können den selben Wert mehrmals ziehen) aus den verfügbaren Trainingsdaten gezogen werden.

Funktionsweise Bagging:

1.   Erzeugen/Ziehen von sehr vielen zufälligen Datensätzen (Subsamples)       aus einem vorhandenen Datensatz (mit Zurücklegen).
2.   Trainieren von geeigneten Bagging-Algorithmen, z.B. CART-Algorithmen, je Subsample.
3.   Durchschnitt der Vorhersagen aller Modelle bilden.

Beispiel: Wir bauen einen Ensemle-Algorithmus mit 5 Entscheidungsbäumen. Diese Entscheidungsbäume machen die folgenden Vorhersagen: Blau, Blau, Rot, Blau, Rot. Die am frequentesten vorausgesagte Klasse (Blau), wird als finale Vorhersage genutzt. So können hohe Varianzen, die typisch für Entscheidungsbäume sind, ausgeglichen werden. Zudem spielt ein mögliches Overfitting einzelner Bäume eine untergeordnete Rolle.

<u>**Boosting:**</u>

Boosting ist ein Vorgang, der beim Machine Learning verwendet wird, um Fehler bei der prädiktiven Datenanalyse zu reduzieren. Das Boosting verbessert die Vorhersagegenauigkeit und die Leistung von Machine Learning Modellen, indem es mehrere schwache Lerner in ein einziges starkes Lernmodell umwandelt. Dazu wird ein Modell aus den Trainingsdaten erstellt und dann ein zweites Modell erstellt, das versucht, die Fehler des ersten Modells zu korrigieren. Es werden so lange Modelle hinzugefügt, bis der Trainingssatz perfekt vorhergesagt ist oder eine maximale Anzahl von Modellen hinzugefügt wurde. Machine Learning Modelle können *schwache* oder *starke Lerner* sein:

<u>Schwache Lerner (Weak Learner):</u>
Schwache Lerner haben eine geringe Vorhersagegenauigkeit (etwas besser als Zufall), ähnlich wie beim zufälligen Raten. Sie neigen zur Überanpassung, d. h. sie können keine Daten klassifizieren, die zu sehr von ihrem ursprünglichen Datensatz abweichen. Wenn Sie das Modell zum Beispiel darauf trainieren, Katzen als Tiere mit spitzen Ohren zu erkennen, kann es sein, dass es eine Katze mit gekrümmten Ohren nicht erkennt. Bsp: Entscheidungsbaum mit einer Ebene, aber auch KNN (k=1) oder Naive Bayes (mit einer Input-Variablen) etc. werden in diesem Kontext oft als schwache Lerner bezeichnet.

<u>Starke Lerner (Strong Learner):</u>
Starke Lerner haben eine höhere Vorhersagegenauigkeit. Durch das Boosting wird ein System schwacher Lerner in ein einziges starkes Lernsystem umgewandelt. Um beispielsweise das Bild einer Katze zu identifizieren, werden ein schwacher Lerner, der spitze Ohren erkennt, und ein weiterer Lerner, der katzenförmige Augen erkennt, kombiniert. Nachdem das Bild des Tieres auf spitze Ohren untersucht wurde, analysiert das System es noch einmal auf katzenförmige Augen. Dadurch wird die insgesamte Genauigkeit des Systems verbessert.

Beispiele für Boosting-Algorithmen sind AdaBoost (Adaptive Boosting), Gradient Boosting oder XGBoost (Extreme Gradient Boosting).

<u>**Bagging vs Boosting**</u>

Das Boosting und das Bagging sind die beiden gebräuchlichen Ensemble-Methoden, mit denen die Vorhersagegenauigkeit verbessert werden kann. Der Hauptunterschied zwischen diesen Lernmethoden ist die Trainingsart. Beim Bagging wird die Genauigkeit von schwachen Lernern verbessert, indem sie mehrere davon gleichzeitig auf mehreren Datensätzen trainieren (Parallele Verabreitung). Im Gegensatz dazu werden beim Boosting schwache Lerner nacheinander trainiert (Sequentielle Verarbeitung).

**2.1 Random Forest** (Bagging)

Random Forest ist ein Ensemble aus Decision Tree Algorithmen und stellt eine Verbesserung gegenüber Bagged Decision Trees dar. Ein Problem mit Entscheidungsbäumen wie CART ist, dass sie *greedy* sind. Daher können die Entscheidungsbäume selbst beim Bagging viele strukturelle Ähnlichkeiten aufweisen, was wiederum zu einer hohen Korrelation ihrer Vorhersagen führt. Die Kombination von Vorhersagen aus mehreren Modellen in Ensembles funktioniert besser, wenn die Vorhersagen aus den Teilmodellen unkorreliert oder bestenfalls schwach korreliert sind. Random Forest ändert den Algorithmus für die Art und Weise, wie die Teilbäume gelernt werden, so dass die resultierenden Vorhersagen aus allen Teilbäumen eine geringere Korrelation aufweisen.

<u>Imbalanced Classification:</u>

<u>K-Fold Cross Validation:</u> Die beiden gebräuchlichsten Ansätze für die Modellbewertung sind der Train-/Test Split und die K-Fold Cross Validation.

Vorgehensweise:

1.   Datensatz (zufällig) mischen (nicht bei Zeitreihen)
2.   Datensatz in k Gruppen aufteilen (z.B. k=5)
3.   Ein Teil der Daten wird als Hold-Out genutzt auf denen das Modell getestet wird; Der restliche Teil der Daten wird für das Training des Modells verwendet
4.   Trainieren und bewerten des Modells
5.   Wiederholen der Schritte 3,4,5

Stratified K-Fold Cross-Validation wird für unbalancierte Datensätze genutzt, sodass das Verhältnis der Klassen in den einzlenen Folds bestehen bleibt.

Beispiel eines Random Forest Algorithmus für eine Klassifikation

```python
from sklearn.datasets import make_classification

# Generieren eines DS
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=3)

print(X.shape, y.shape)
```

```python
# Evaluation eines RF-Classifiers mit K-Fold CV
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier

# Initialisieren des Modells
model = RandomForestClassifier() # Default von 100 Bäumen

# Evaluieren des Modells
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')

# Performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

Wie wirkt sich die Anzahl der Bäume auf die Klassifikationsgenauigkeit aus?

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot

# Datensatz erzeugen
def get_dataset():
	X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=3)
	return X, y

# Erzeugen einer Liste von Modellen, die evaluiert werden sollen
def get_models():
	models = dict()
	# Definieren der Anzahl an Bäumen
	n_trees = [10, 50, 100, 500, 1000]
	for n in n_trees:
		models[str(n)] = RandomForestClassifier(n_estimators=n)
	return models

# Evaluation mittels CV
def evaluate_model(model, X, y):
	cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
	scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
	return scores

# Datensatz erzeugen
X, y = get_dataset()
# Modell(e) initialisieren
models = get_models()
# Modelle evaluieren
results, names = list(), list()
for name, model in models.items():
	scores = evaluate_model(model, X, y)
	results.append(scores)
	names.append(name)
	# Performance der einzelnen Modelle
	print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
# Visualisierung der Ergebnisse in einem Boxplot
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.show()
```

**2.2 AdaBoost** (Boosting)

AdaBoost (Adaptive Boosting) ist eine Variante des Boosting Algorithmus und kann für binäre Klassifikationsprobleme auf Basis von Entscheidungsbäumen genutzt werden. Ein AdaBoost-Klassifikator ist ein Meta-Estimator, der zunächst einen Klassifikator auf den ursprünglichen Datensatz fittet und dann weitere Kopien des Klassifikators auf denselben Datensatz fittet, wobei jedoch die Gewichte der falsch klassifizierten Instanzen so angepasst werden, dass sich die nachfolgenden Klassifikatoren stärker auf schwierige Fälle konzentrieren.

<u>Meta-Estimator:</u> Ist ein Modell, dass die Parameter eines Vorhänger-Modells als Input nutzt.

grafik.png

```python
import numpy as np
import pandas as pd

# Datensatz einlesen
iris = pd.read_csv('Iris.csv')
```

```python
# Kopf des DataFrames ausgeben
iris.head()
```

```python
# Zusammenfassung zum Datensatz
iris.info()
```

```python
# Features in eine Variable überführen
X = iris.drop('Species', 1)

X.head()
```

```python
# Label in Variable hinterlegen
y = iris['Species']

y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label kodieren
le=LabelEncoder()

y=le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split

# Train-/Test-Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Model initiaisieren
model = AdaBoostClassifier(n_estimators=50, learning_rate=1, random_state=0)

# Model trainieren
model.fit(X_train, y_train)

# Prädiktion durchführen
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Genauigkeit des Modells bestimmen
print("AdaBoost Classifier Model Accuracy:", accuracy_score(y_test, y_pred))
```

**AdaBoost vs RandomForest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Model initiaisieren
model = RandomForestClassifier()

# Model trainieren
model.fit(X_train, y_train)

# Prädiktion durchführen
y_pred = model.predict(X_test)

# Genauigkeit des Modells bestimmen
print("Random Forest Classifier Model Accuracy:", accuracy_score(y_test, y_pred))
```
