**Äpfel mit Orangen (nicht mit Birnen) vergleichen!**
<br>
Welche Art von System ermöglicht es uns, Äpfel und Orangen zu unterscheiden, ohne eine Vielzahl von Regeln zu definieren?
<br>
<br>
Ein **Classifier**!
<br>
Vereinfacht gesagt, ist ein Classifier eine Funktion, die Daten (z.B. Textur, Gewicht, Farbe) als Input nimmt und ein Label ausgibt (z.B. Orange). Ein bekanntes Beispiel ist die Klassifizierung von Mails in normale Mails und Spam. 
<br>
<br>
Aber wie können wir ein solchen Classifier anlernen? Über ein **Supervised Learning** Verfahren! Das heißt, wir zeigen dem Classifier Beispiele von dem Problem, das er später lösen soll und davon möglichst viele:
<br>
<br>
**Beispiel**: Gewicht: 250g; Textur: Glatt --> Label: Apfel


Unsere Vorgehensweise: 
<br>
1. Daten sammeln & aufbereiten
2. Classifier trainieren
3. Vorhersagen machen

```python
#Importieren der sklearn Bibliothek
#sklearn ist eine Bibliothek für Machine Learning und bietet verschiede Algorithmen für Klassifikation, Regression und Clustering
from sklearn import tree

#Schritt 1: Daten sammlen
#Wir benötigen Features, also Merkmalsausprägungen der jeweiligen Klasse, und Lables als Zielvariable
#In diesem Beispiel nutzen wir für die Definition von Features und Label wieder Arrays, wie wir es bereits in NumPy kennengelernt haben.

#Frage: Was muss noch getan werden, bevor wir die Daten dem Classifier zum trainieren übergeben können?

features = [[140, "smooth"], [130, "smooth"], [150, "bumpy"], [170, "bumpy"]]

labels = ["apple", "apple", "orange", "orange"]


#Schritt 2: Classifier trainieren
#DecisionTreeClassifier: Ermöglicht MultiClass-Klassifikationen
clf = tree.DecisionTreeClassifier()

#Wir übergeben dem Classifier 2 Arrays: Features und Labels
clf = clf.fit(features, labels)


#Schritt 3: Vorhersagen machen
print(clf.predict([[160, 1]]))
```
